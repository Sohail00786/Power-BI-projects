## Task 1 - Data Cleaning

```python
# Library for Data Modeling and Cleaning
import pandas as pd
```

```python
# Importing Data
content = pd.read_csv('Content.csv')
reactions =pd.read_csv('Reactions.csv')
reaction_types = pd.read_csv('ReactionTypes.csv')
```

```python
# for upper most columns
content.head()
```

```python
reactions.head()
```

```python
reaction_types.head()
```

```python
# Dropping Unwanted Columns
content.drop('Unnamed: 0', axis = 1, inplace = True)
reactions.drop(['Unnamed: 0','User ID'], axis = 1, inplace = True)
reaction_types.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
# merging reaction types with reactions 
join_1 =pd.merge(reaction_types, reactions, on = ['Type'], how = 'inner')
```

```python
# Upper Columns of Merged Dataset (reaction and reaction Types)
join_1.head()
```

```python
# Merging join_1 with content
join_2 =pd.merge(join_1, content, on = ['Content ID'], how = 'inner')
```

```python
# # Upper Columns of Merged Dataset( join_1 and Content)
join_2.head()
```

```python
# Storing final data set in variable for further analysis
final = join_2
```

```python
# data information
final.info()
```

```python
final.isnull().sum()
```

```python
# Dropping URL
final.drop(['URL'], axis = 1, inplace = True)
```

```python
final.head()
```

```python
final['Type_x'].unique()
```

```python
final['Category'].unique()
```

```python
# Converting to lower to address column for categorical variable
final['Category'] = final['Category'].replace('"', '', regex = True)
final['Category'] = final['Category'].str.lower()
```

```python
final['Category'].unique()
```

```python
final.head(6)
```

```python
final.rename({'Type_x':'Reaction'}, axis = 1, inplace = True)
final.rename({'Type_y':'Type'}, axis = 1, inplace = True)
```

```python
final.head()
```

```python
# exporting Merged CSV
final.to_csv('clean.csv', index = False)
```
